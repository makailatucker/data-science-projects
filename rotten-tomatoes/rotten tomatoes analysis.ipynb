```python
import pandas as pd
import numpy as np

all_films = pd.read_csv('all_films_wayback_reviews.csv')
all_films
```

```python
all_films = pd.concat([all_films, pd.read_csv('all_films_wayback_2019_reviews.csv'), pd.read_csv('all_films_wayback_reviews_2022.csv'), pd.read_csv('all_films_wayback_aqua_reviews.csv')])
```

```python
all_films['critic'] = pd.to_numeric(all_films['critic'].str.replace('%', '').replace('Tomatometer', '').replace("error", ''))
all_films['audience'] = pd.to_numeric(all_films['audience'].str.replace('%', '').replace('Coming', '').replace('Audience', '').replace("error", ''))
all_films
```

```python
all_films[all_films["film"] == "Aquaman"]
all_films["audienceNum"] = all_films["audienceNum"].str.replace(',', '').replace('\+ Ratings', '')
```

```python
all_films["audienceNum"] = all_films["audienceNum"].str.replace('\+ Ratings', '')
all_films["audienceNum"] = all_films["audienceNum"].str.replace('User Ratings: ', '').replace("Not yet available", '')
all_films["audienceNum"] = all_films["audienceNum"].str.replace('Ratings: Not yet available', '')
```

```python
# all_films["audienceNum"].unique()
all_films['criticNum'] = pd.to_numeric(all_films['criticNum'].str.replace(' Reviews', ''))
all_films['audienceNum'] = pd.to_numeric(all_films['audienceNum'].str.replace('User Ratings\: ', '').replace('liked it', '').replace('Coming', '').replace('coming', '').replace("Aquaman", '').replace("Star Wars: The Last Jedi", ''))

```

```python
import matplotlib.pyplot as plt

def plot_film(film):
    df = all_films[all_films['film']==film]
    series = df[['critic', 'audience','date']].set_index('date')
#     print(df)
    plt.plot(series)
    plt.axhline(y=75, color='yellow', linestyle='dashed',linewidth=0.5)
    plt.axhline(y=60, color='red', linestyle='dashed',linewidth=0.5)
    plt.axhspan(75, 100, alpha=0.5, color='lightgreen')
    plt.axhspan(60,75, alpha=0.5, color='#F6F7A8')
    plt.axhspan(0, 60, alpha=0.5, color='salmon')    
    plt.ylim(max(min(series.min().values)-5, 0), min(max(series.max().values)+5,100))
#     print(series.index[::10])
    plt.xticks(series.index[::10],rotation=45, ha="right")
    plt.locator_params(axis='x', nbins=10)
    plt.title(film + " Rotten Tomatoes Score over Time")
```

```python
def plot_reviews(film):
    df = all_films[all_films['film']==film]
    series = df[['criticNum', 'audienceNum', 'date']].set_index('date')[:60]
    plt.plot(series)
    plt.xticks(series.index[::10],rotation=45, ha="right")
    plt.locator_params(axis='x', nbins=10)
    plt.title("Number of " + film + " Reviews over Time")
```
